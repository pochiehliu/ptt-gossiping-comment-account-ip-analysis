## Q1: Did author push own post?

```python
temp = pd.merge(
    df_comment[['postID', 'commetAuthor','Indicator']] \
        .assign(postID=df_comment.postID.astype(str)),
    df_post[['postID','Author']]. \
        assign(postID=df_post.postID.astype(str)),
    how='left', 
    on='postID')

temp = temp[temp['Author']==temp['commetAuthor']] \
    .groupby(['postID','Author','Indicator']) \
    .size().reset_index(name='counts')

temp.sort_values(by = 'counts', ascending = False)
```

Based on the indicator type, all authors' comments are arrows.

## Q2: Did author use other accounts to push own post?

```python
temp = pd.merge(
    df_comment[['postID', 'commetAuthor','Indicator','commentIP']] \
        .assign(postID=df_comment.postID.astype(str)),
    df_post[['postID','Author','IP']] \
        .assign(postID=df_post.postID.astype(str)),
    how='left', 
    on='postID')

temp = temp[(temp['commentIP']==temp['IP']) &
            (temp['commetAuthor']!=temp['Author'])
           ].groupby(['postID','Author','commetAuthor','Indicator','IP']) \
            .size().reset_index(name='counts')

temp = temp[temp['Indicator']!=3].sort_values(by = 'counts', ascending = False)
temp
```

## Q3: Can we find same people or group users?

```python
temp = df_comment[['commentIP','commetAuthor']] \
       .drop_duplicates().groupby(['commentIP']) \
       .size().reset_index(name='counts') \
       .sort_values(by = 'counts', ascending = False)
temp = temp[temp.counts>4]
temp
```

```python
temp = pd.merge(
    df_comment[['commetAuthor','commentIP']] \
        .assign(commentIP=df_comment.commentIP.astype(str)),
    df_comment[['commetAuthor','commentIP']] \
        .assign(commentIP=df_comment.commentIP.astype(str)),
    how='left', 
    on='commentIP').drop_duplicates()
```

```python
temp = temp[temp['commetAuthor_x']!=temp['commetAuthor_y']][['commetAuthor_x','commetAuthor_y']].drop_duplicates()
temp.columns = ['from','to']
```

```python
import networkx as nx
import hvplot.networkx as hvnx
import holoviews as hv
```

```python
G=nx.from_pandas_edgelist(temp, 'from', 'to')
```

```python
hvnx.draw(G, nx.spring_layout(G),
          with_labels=True,
          height = 800, width = 800)
```

```python
graphs = list(nx.connected_component_subgraphs(G))
sub_graphs = [g for g in graphs if g.number_of_nodes() > 2]
sub_graphs.sort(key = lambda g: g.number_of_nodes(),reverse=True)
```
